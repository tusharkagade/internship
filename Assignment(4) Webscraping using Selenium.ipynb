```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.naukri.com/')
```

```python
designation=driver.find_element(By.CLASS_NAME,'suggestor-input')
designation.send_keys('Data Scientist')
```

```python
location=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input')
location.send_keys('Delhi/NCR')
```

```python
search=driver.find_element(By.CLASS_NAME,'qsbSubmit')
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
locat=[]
locat1=[]
```

```python
title_tags=driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in title_tags:
    title=i.text
    job_title.append(title)
    
location_tags=driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-location loc"]')
for i in location_tags:
    location=i.text
    job_location.append(location)

company_tags=driver.find_elements(By.XPATH,'//div[@class=" row2"]/span/a[1]')
for i in company_tags:
    company=i.text
    company_name.append(company)
    
exper_tags=driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
for i in exper_tags:
    exper=i.text
    experience_required.append(exper)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
data=pd.DataFrame({'title':job_title,'location':job_location,'company_name':company_name,'experience':experience_required})
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1')
```

```python
Name=[]
Born_Dead=[]
Term_of_office=[]
Remarks=[]
```

```python
name_tags=driver.find_elements(By.XPATH,'//td[@valign="top"]/div')
for i in name_tags:
    name_title=i.text
    Name.append(name_title)
    
    
born_tags=driver.find_elements(By.XPATH,'//td[@valign="top"]/div')
for i in born_tags:
    born_title=i.text
    Born_Dead.append(born_title)
    
    
    
term_tags=driver.find_elements(By.XPATH,'//td[@valign="top"]/div')
for i in term_tags:
    term_title=i.text
    Term_of_office.append(term_title)
    
    
    
    
remarks_tags=driver.find_elements(By.XPATH,'//td[@valign="top"]/div')
for i in remarks_tags:
    remarks_title=i.text
    Remarks.append(remarks_title)
```

```python
print(len(Name),len(Born_Dead),len(Term_of_office),len(Remarks))
```
